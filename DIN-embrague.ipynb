# Ejercicio Embrague

![title](E-embP_13-71.png)

>Suponga que el sistema trabaja de manera horizontal. La fricción de las zapatas A contra el marco B hace funcional el embrague centrífugo y se transmite el torque de A hacia B. La fuerza centrífuga mantiene a las zapatas contra el marco B donde su coeficiente de fricción estático es de $\mu=0.8$. Cada zapata pesa 9.8[N] y el peso del resto de partes es despreciable. 
- Calcular la tensión inicial de los resortes (C) para que el contacto de las zapatas se haga cuando $\dot{\theta}\geq200[rpm]$
- Determinar el torque en el intervalo $100\geq\dot{\theta}\geq1500[rpm]$

# Sistema de referencia
## Coordenadas radial-transversal

> Se realiza el diagrama de cuerpo libre representando todas las fuerzas. (Foto adjuntada)
> Las zapatas (A) se van acercando hacia la carcasa (B) conforme esta gira en sentido horario, el contacto de estos dos se produce solamente cuando $\dot{\theta}\geq200[rpm]$. 




> Realizamos la sumatoria de fuerzas en el eje radial, donde se obtuvo lo siguiente:

$$\Sigma F e_{r}=m\cdot a_{r}$$

> La aceleración radial está compuesta de una parte centrípeta y de otra radial, es decir:
$$a_{r} = -r\cdot\dot{\theta}^2 + \ddot{r}$$

> Debido a que $r$ tiene un límite al llegar a la carcasa B, y mantiene un valor de velocidad constante.
> Se obtiene entonces las siguientes fuerzas en la sumatoria.

$$ F_{e}+N=r\cdot\dot{\theta}^2 \tag{1} \label{F_{e}}$$

 > Despejando la $F_{e}$ se tiene,
$$F_{e}=r\cdot\dot{\theta}^2-N$$

> Se desea obtener la tensión inicial en el resorte tal que el contacto entre la zapata y la carcasa ocurra cuando$\dot{\theta}\geq200[rpm]$, en donde la $N=0$.
$$F_{e}=r\cdot\dot{\theta}^2$$
$$F_{e}=0.3\cdot\left(\frac{200\cdot2\pi}{60}\right)^2$$
$$F_{e}=131.59N$$

> Para la parte b) se desea hallar el torque en el intervalo $100\leq\dot{\theta}\leq1500[rpm]$, para esto realizamos la sumatoria en el eje transversal.

$$\Sigma F e_{\theta}=m\cdot a_{\theta}$$

> La aceleración transversal está compuesta de una parte de coriolis y otra de euler, es decir:

$$a_{\theta} = 2\dot{r}\dot{\theta}+r\ddot{\theta}$$

> Debido a que las zapatas giran con una $\dot{\theta}=cte$ (asumido), se tiene que $\ddot{\theta}=0$.
> Se obtiene entonces las siguientes fuerzas en la sumatoria.

$$ -F_{r_{e}}=m\cdot 2\dot{r}\dot{\theta}$$

$$ F_{r_{e}}=-m\cdot 2\dot{r}\dot{\theta}$$

> Por otro lado, despejando la fuerza $N$ en función de $\dot{\theta}$, de la ecuación (1).

$$N=m\cdot r\cdot\dot{\theta}^2-F_{e}$$

> Calculamos la fuerza Normal para $\dot{\theta}=1500[rpm]$ y $F_{e}=131.59N$

$$N=1\cdot 0.3\cdot\left(\frac{1500\cdot2\pi}{60}\right)^2-131.59$$

$$N=7270.61[N]$$

> Calculamos la $F_{r_{e}}$,

$$F_{r_{e}}=\mu\cdot N$$

$$F_{r_{e}}=0.8 \cdot7270.61$$

$$F_{r_{e}}=5816.49N$$

> Calculamos el torque con la $F_{r_{e}}$ encontrada y sabiendo que para el sistema existirán dos fuerzas que realizan el torque total:

$$\tau=F_{r_{e}}\cdot 2\cdot r_{f}$$

$$\tau=5816.49\cdot 2\cdot 0.4$$

$$\tau=4653.19 [Nm]$$

> Para la $\dot{\theta}=100[rpm]$ la fuerza normal no existe, ya que no hay contacto con la carcasa por lo que el torque para esa velocidad angular es:

$$\tau=0 [Nm]$$

>Por lo tanto, el intervalo el intervalo $100\leq\dot{\theta}\leq1500[rpm]$ es:
$$0\leq\tau\leq4653.19[Nm]$$
