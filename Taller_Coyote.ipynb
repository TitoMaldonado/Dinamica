# Taller Coyote
>Considere que se modela el sistema de la piedra que cae de la siguiente manera (2D): <br/>
 <br/>
Un cuerpo cuadrado de b=3[m] de masa m=2000[kg] que pivota sobre la esquina (punto O). Suponga que la piedra rueda sin deslizar y que hay una fricción en la rueda con µ=0.3
>Calcule
- el ángulo al cuál el cuerpo pierde contacto con la esquina (punto O) y empieza a caer.
- Se necesitará realizar un gráfico de la variación de la normal con respecto al ángulo de giro
- realice un gráfico de la variación de la velocidad angular hasta el mismo ángulo del literal anterior
- ¿podría pasar lo que pasó en el vídeo?

Se realiza el DCL del cuerpo cuadrado: <br/>


Se tiene que la inercia de la sección rectangular alrededor de O es:
$I_o=\frac{1}{12}m(b^2+b^2)+m(\frac{b}{2})^2$ <br/> Se tiene entonces que $I_o=\frac{5}{12}mb^2$ = $I_o=\frac{5}{12}mb^2$= $I_o=7500$ $[kg*m^2]$

Se realiza la sumatoria de momentos y se tiene: <br/>
$\sum M_o=I_o\alpha$ <br/>
$mg\frac{b}{2}\sin\theta=\frac{5}{12}mb^2\alpha$ <br/>
Se despeja $\alpha$ y se tiene que $\alpha=\frac{6}{5}\frac{g}{b}\sin\theta$ <br/>
$\alpha=\sin\theta$

Se realiza la integral $\int_{0}^{\omega}\!\omega \,d\omega=\int_{0}^{\theta}\!\alpha \,d\theta$ <br/>
$\omega^2=\frac{12}{5}\frac{g}{b}\int_{0}^{\theta}\!\sin\theta \,d\theta$ <br/>
$\omega^2=\frac{12}{5}\frac{g}{b}(1-\cos\theta)$ <br/>
$\omega^2=\frac{12}{5}\frac{g}{3}(1-\cos\theta)$ <br/>
$\omega^2=7,848(1-\cos\theta)$ <br/>
$\omega=\sqrt{7,848(1-\cos\theta)}$

Se realiza la suamtoria de fuerzas en el eje normal <br/>
$\sum F_n=ma_n$ <br/>
$mg\cos\theta-N=m\frac{b}{2}\frac{12}{5}\frac{g}{b}(1-\cos\theta)$ <br/>
$N=\frac{mg}{5}(11\cos\theta-6)$ <br/>
$N=3924(11\cos\theta-6)$

Se realiza la suamtoria de fuerzas en el eje tangencial <br/>
$\sum F_t=ma_t$ <br/>
$mg\sin\theta-Fr=m\frac{b}{2}\frac{6}{5}\frac{g}{b}\sin\theta$ <br/>
$Fr=\frac{2}{5}mg\sin\theta$ <br/>
$Fr=7848\sin\theta$

El ángulo al cuál el cuerpo comenzará a caer se determinará por: <br/>
Se tiene que $Fr=\mu N$ <br/>
$Fr=0,3N$ <br/>
$\frac{2}{5}mg\sin\theta$=$0,3\frac{mg}{5}(11\cos\theta-6)$ <br/>
$\sin\theta$=$\frac{3}{20}(11\cos\theta-6)$ <br/>
$\theta=31°$

```python
th=np.arange(0,180,1)
vth = np.array(np.deg2rad(th))
def f1(th):
    return 3924*(11*np.cos(vth)-6)
fig=go.Figure()
fig.add_trace(go.Scatter(x=th, y=f1(th), 
                         mode='lines', 
                         name='funcion'))
fig.update_layout(title='normal en función de la variación del ángulo de rotación',
                   xaxis_title='grados',
                   yaxis_title='N')

fig.show()
```
